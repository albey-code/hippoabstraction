## Plotting only the 9 LH ROIs

```python
from nilearn import plotting, image, datasets
import os
```

```python
# Indices for LH ROIs (see roi_filenames above)

lh_filenames = [
    "roi_1_x35_y34_z15.nii.gz",  # LH ROIs based on plotting
    "roi_2_x35_y33_z16.nii.gz",
    "roi_3_x35_y32_z17.nii.gz",
    "roi_4_x35_y31_z18.nii.gz",
    "roi_5_x35_y29_z19.nii.gz",
    "roi_6_x35_y28_z20.nii.gz",
    "roi_7_x35_y27_z21.nii.gz",
    "roi_8_x35_y26_z22.nii.gz",
    "roi_9_x35_y25_z23.nii.gz",
]
```

```python
# Full paths to each ROI file
lh_roi_paths = [os.path.join(base_path, fname) for fname in lh_filenames]
```

```python
# Load MNI152 template
mni = datasets.load_mni152_template()
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
```

```python
# Loop over each left hippocampal ROI
for i, roi_path in enumerate(lh_roi_paths, 1):  # Start index at 1 for ROI labeling
    plotting.plot_roi(
        roi_path,
        bg_img=mni,
        title=f"ROI {i} - LH",
        cmap="autumn",
        alpha=0.9,
        threshold=0.1,
        display_mode='ortho',
        draw_cross=False
    )
```

```python
# Load all LH ROI images
roi_imgs = [image.load_img(p) for p in lh_roi_paths]

# Create a single 4D image by stacking along a new axis
roi_4d = image.concat_imgs(roi_imgs)

# Merge into one mask (i.e. any voxel that is part of any ROI)
roi_combined = image.math_img("np.max(img, axis=-1)", img=roi_4d)

# Plot the combined ROIs on the MNI template
plotting.plot_roi(
    roi_combined,
    bg_img=mni,
    title="All Left Hippocampal ROIs (Fine Gradient)",
    cmap="autumn",
    alpha=0.9,
    threshold=0.1,
    display_mode='ortho',
    draw_cross=False
)
```
