```python
from google.colab import files
import scipy.io
import numpy as np
```

```python
# Upload file(s)
uploaded = files.upload()
```

The shape is probably the timepoints (per run) x the regressors. Here, _, 23 arguably does not mean the number of N = 23 participants, but the different regressors, e.g., motion, physio and nuisance etc.

```python
# Load and inspect contents
for filename in uploaded.keys():
    print(f"\nInspecting {filename}")
    mat = scipy.io.loadmat(filename)
    for key in mat:
        if not key.startswith('__') and isinstance(mat[key], np.ndarray):
            print(f"Variable '{key}' found with shape {mat[key].shape}")
```

Excluding the first 5 rows (in the same way that the first 5 volumes were excluded for the fMRI data) to match the event .tsv files. And because: "The first five volumes of each block were discarded to allow for scanner equilibration." from Garvert et al. (2017).

```python
# All 23 regressors (motion + physio)
motion1 = scipy.io.loadmat('23_R_and_rp_session1_slice15.mat')['R']
motion2 = scipy.io.loadmat('23_R_and_rp_session2_slice15.mat')['R']
motion3 = scipy.io.loadmat('23_R_and_rp_session3_slice15.mat')['R']
```

```python
import pandas as pd
```

```python
# Save each regressor matrix as .tsv
pd.DataFrame(motion1).to_csv('sub-23_run-1_motion_physio.tsv', sep='\t', index=False, header=False)
pd.DataFrame(motion2).to_csv('sub-23_run-2_motion_physio.tsv', sep='\t', index=False, header=False)
pd.DataFrame(motion3).to_csv('sub-23_run-3_motion_physio.tsv', sep='\t', index=False, header=False)
```
