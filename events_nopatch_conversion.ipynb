This notebook converts the event files containing .txt files for all objects with and without a patch into **event files merely containing objects without a patch.**

Load event files for 23 subjects and 3 runs, remove files containing "patch" in their names, and provide the remaining files for download.

```python
from google.colab import files
import zipfile, os, glob, io
import pandas as pd
```

```python
uploaded = files.upload()
```

```python
# Get the name dynamically
zip_filename = list(uploaded.keys())[0]
```

```python
# Unzip
unzip_dir = "unzipped_run1"
# Clear the unzip directory if it exists
if os.path.exists(unzip_dir):
    for root, dirs, files in os.walk(unzip_dir, topdown=False):
        for name in files:
            os.remove(os.path.join(root, name))
        for name in dirs:
            os.rmdir(os.path.join(root, name))
os.makedirs(unzip_dir, exist_ok=True)
with zipfile.ZipFile(zip_filename, 'r') as zip_ref:
    zip_ref.extractall(unzip_dir)

print(os.listdir(unzip_dir))  # Sanity check
```

```python
# Print number of entries in unzip_dir
print(len(os.listdir(unzip_dir))) # Check it's 15
```

```python
# Filter out files ending with 'patch_1.txt'
filtered_files = [f for f in os.listdir(unzip_dir) if not f.endswith('patch_1.txt')]

print("Files after filtering:")
print(filtered_files)
```

```python
# Create a new zip file containing the filtered 'patch_0.txt' files
zip_filename_nopatch = "subject23_run3_np.zip"
with zipfile.ZipFile(zip_filename_nopatch, 'w') as zipf:
    for file_name in filtered_files:
        file_path = os.path.join(unzip_dir, file_name)
        zipf.write(file_path, os.path.basename(file_path))

print(f"Created new zip file: {zip_filename_nopatch}")
```
