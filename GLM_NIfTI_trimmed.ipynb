Step 1: Load fMRI NIfTI files

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import nibabel as nib
```

```python
base_path = '/content/drive/MyDrive/Mona_Garvert_2017_fMRI_Data/'
```

```python
img1 = nib.load(base_path + 'sub04_run1_bold.nii.gz')
img2 = nib.load(base_path + 'sub04_run2_bold.nii.gz')
img3 = nib.load(base_path + 'sub04_run3_bold.nii.gz')
```

```python
print("Run 1 shape:", img1.shape)
print("Run 2 shape:", img2.shape)
print("Run 3 shape:", img3.shape)
```

Step 2: Trim first 5 volumes of the NIfTI files

```python
import os
```

```python
# Output directory to save the trimmed files
output_dir = '/content/drive/MyDrive/Mona_Garvert_2017_fMRI_Data/trimmed/'
os.makedirs(output_dir, exist_ok=True)
```

```python
# Trim and save each run
trimmed_img1 = nib.Nifti1Image(img1.get_fdata()[..., 5:], img1.affine, img1.header)
nib.save(trimmed_img1, os.path.join(output_dir, 'sub-04_run-1_bold_trimmed.nii.gz'))

trimmed_img2 = nib.Nifti1Image(img2.get_fdata()[..., 5:], img2.affine, img2.header)
nib.save(trimmed_img2, os.path.join(output_dir, 'sub-04_run-2_bold_trimmed.nii.gz'))

trimmed_img3 = nib.Nifti1Image(img3.get_fdata()[..., 5:], img3.affine, img3.header)
nib.save(trimmed_img3, os.path.join(output_dir, 'sub-04_run-3_bold_trimmed.nii.gz'))
```

```python
# Confirm new shapes
print("Trimmed Run 1 shape:", trimmed_img1.shape)
print("Trimmed Run 2 shape:", trimmed_img2.shape)
print("Trimmed Run 3 shape:", trimmed_img3.shape)
```
