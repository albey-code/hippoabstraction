````python
from scipy.io import loadmat
import numpy as np

# Load the .mat file with helpful options
mat = loadmat("scan_1.mat", squeeze_me=True, struct_as_record=False)
data = mat["data"]

# Get train as a flat array of blocks
train = np.atleast_1d(data.train).ravel()
print("num training blocks:", len(train))
````

Substitute the `first_train_block = train[0]` with numbers from 0 to 18 (for the 19 training blocks). Beware, `train[10]` and `train[11]` look odd, there seems to be data missing!!

````python
# Access the first training block
first_train_block = train[0]

# Iterate through the fields and print their content
print("Contents of fields in the first training block:")
for field_name in [f for f in dir(first_train_block) if not f.startswith("_")]:
    try:
        field_content = getattr(first_train_block, field_name)
        print(f"\n--- Field: {field_name} ---")
        # Print the content.
        if isinstance(field_content, np.ndarray):
            print("type/shape:", type(field_content), getattr(field_content, "shape", None))
            # Print the full content of the array
            print("content:", np.asarray(field_content).squeeze())
        else:
            print("content:", field_content)

    except AttributeError:
        print(f"Could not access field: {field_name}")
````

## Allowed object sequence

Check that the `seq` of object indices is valid according to the following convention of one node (object) following the next!

```
graph = [];
graph(1,[2,3,4])         = 1;
graph(2,[1,4,5])         = 1;
graph(3,[1,4,6,7])       = 1;
graph(4,[1,2,3,5,7])     = 1;
graph(5,[2,4,8,9])       = 1;
graph(6,[3,7,10])        = 1;
graph(7,[3,4,6,8,10,11]) = 1;
graph(8,[5,7,9,11,12])   = 1;
graph(9,[5,8,12])        = 1;
graph(10,[6,7,11])       = 1;
graph(11,[7,8,10,12])    = 1;
graph(12,[8,9,11])       = 1;
```
You can read this as: object 1 connects to objects 2, 3, 4 etc.
